```python
def age_group(age):
    if age == 0:
        return "Not borned"
    elif age <= 18:
        return "Child"
    elif age <= 50:
        return "Adult"
    else:
        return "Senior Citizen"
```

```python
df['Age Category']=df['age'].apply(age_group)
df.head()
```

```python
# Which age group has the highest number of admissions?  #
```

```python
df.groupby('Age Category')[['patient_id']].count()
```

```python
# Create cost category feature
#treatment_cost < 5000 then 'low', treatment_cost < 100000 then ' Medium' , else 'High
```

```python
def cost_category(treatment_cost):
    if treatment_cost < 5000:
        return "low"
    elif treatment_cost < 100000:
        return "Medium"
    else:
        return "High"
```

```python
df['cost_category']=df['treatment_cost'].apply(cost_category)
df.head()
```
